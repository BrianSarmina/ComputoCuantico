```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector, DensityMatrix
import numpy as np
import matplotlib.pyplot as plt
```

## Producto tensorial

```python
mas = np.array([1/np.sqrt(2), 1/np.sqrt(2)]) # Estado en superposición para |+>.
menos = np.array([1/np.sqrt(2), -1/np.sqrt(2)]) # Estado en superposición para |/>.

a = np.array([1, 0])  # Por ejemplo, |0⟩
b = np.array([0, 1])  # Por ejemplo, |1⟩
```

```python
producto_tensorial_masmenos = np.kron(mas, menos) # Producto tensorial +- usando numpy.kron().
producto_tensorial_ab = np.kron(a, b) # Producto tensorial ab.

print("Producto tensorial |+⟩ ⊗ |-⟩:")
print(producto_tensorial_masmenos)

print("Producto tensorial |a⟩ ⊗ |b⟩:")
print(producto_tensorial_ab)
```

## Operador de densidad

```python
### Ejemplo 1 ###
# Definir el estado puro |ψ⟩ = 1/√2(|0⟩ + |1⟩)
state = Statevector([1/np.sqrt(2), 1/np.sqrt(2)])

# Calcular el operador de densidad
density_matrix = DensityMatrix(state)
print("Operador de Densidad para |ψ⟩:")
print(density_matrix)
```

```python
### Ejemplo 2 ###
# Definir los estados |+⟩ y |-⟩
plus = np.array([1/np.sqrt(2), 1/np.sqrt(2)])
minus = np.array([1/np.sqrt(2), -1/np.sqrt(2)])

# Crear los operadores de densidad para |+⟩ y |-⟩
rho_plus = np.outer(plus, plus.conj())
rho_minus = np.outer(minus, minus.conj())

# Calcular la mezcla con probabilidades iguales
rho_mixture = 0.5 * rho_plus + 0.5 * rho_minus
print("Operador de Densidad para la mezcla de |+⟩ y |-⟩:")
print(rho_mixture)
```
